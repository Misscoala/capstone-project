```python
!pip install folium
import folium 
from bs4 import BeautifulSoup
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source,'lxml')
print(soup.title)
```

```python
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
dfs = pd.read_html(tab)
df = dfs[0]
df.head()
```

```python
#Ignore cells with a borough that is Not assigned.
df1 = df[df.Borough !='Not assigned']
df1
```

```python
#More than one neighborhood can exist in one postal code area.
df1.rename(columns={'Postal Code':'PostalCode'}, inplace=True) 
df1
```

```python
df2 = df1.groupby(['PostalCode','Borough'], sort=False).agg(','.join)
df2.reset_index(inplace=True)

#If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned',df2['Borough'], df2['Neighbourhood'])
df2
```

Question 3

Explore and cluster the neighborhoods in Toronto. You can decide to work with only boroughs that contain the word Toronto and then replicate the same analysis we did to the New York City data. It is up to you.

```python
df4 = df3[df3['Borough'].str.contains('Toronto')]
df4
```

```python
map_toronto = folium.Map(location=[43.648426,-79.382280],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
# set number of clusters
kclusters = 5

toronto_clustering = df4.drop(['PostalCode','Borough','Neighbourhood'], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_
```

```python
# create map
map_clusters = folium.Map(location=[43.648426,-79.382280], zoom_start=11)
df4.insert(0, 'Cluster Labels', kmeans.labels_)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighbourhood'], df4['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
